## Abstract

***Pemasaran Produk Deposito Berjangka oleh Bank XYZ*** 

Deposito berjangka adalah investasi berjangka tetap yang mencakup penyetoran uang ke dalam akun di lembaga keuangan. ## Metrics Calculation

Ada 2 metrics yang ingin kita hitung:
- Cost dari marketing
- Profit dari produk deposito berjangka

***Cost Marketing***

Berdasarkan sumber [disini](https://digitalbanking.asee.io/how-much-does-it-cost-for-a-bank-to-acquire-a-net-new-customer/) , ada metric yang digunakan oleh bank dalam menghitung cost penarikan customer. yang biasa disebut Customer Acquisition Cost (CAC) yang didapat dari cost marketing dibagi dengan total customer yang didapat. Dan menurut sumber [disini](https://digitalbanking.asee.io/how-much-does-it-cost-for-a-bank-to-acquire-a-net-new-customer/) kita bisa melihat rata-rata budget yang dikeluarkan oleh bank dalam melakukan marketing. Karena bank XYZ adalah bank buku 3 yang valuasi asetnya diatas 8 trilliun rupiah (500 million dollars),  kita bisa mengambil data berdasarkan sumber, bahwa bank XYZ mengeluarkan budget kurang lebih $387000 . dengan menggunakan perhitungan CAC, kita bisa menghitung cost dari marketing kita dengan cara berikut:

- ***387000 / 3700 OF CUSTOMER IN DATA THAT DEPOSITED THE MONEY  = $105 DOLLAR COST OF MARKETING***

3700 didapat dari jumlah nasabah yang melakukan deposito berjangka dari data yang kita punya. jadi, 

***$105 dollar adalah cost dari melakukan marketing deposito berjangka oleh bank XYZ***

Jadi, Jika ada error dari False Positive (mengkategorikan orang akan mendeposit uang ke bank padahal tidak), harga yang harus dihitung dari tiap error adalah \$105 per nasabah.

<br>

***Profit Deposito Berjangka***

menurut sumber [disini](https://www.westpac.com.au/personal-banking/bank-accounts/term-deposit/savings-vs-term-deposit/#:~:text=Most%20term%20deposits%20will%20have,for%20a%20period%20of%20time), kebanyakan dari program deposito berjangka memiliki minimum deposit per depositor sekitar 1000 sampai 5000 dollar. anggaplah bahwa kemungkinan terburuk, setiap orang yang akan membeli produk deposito ini akan mendeposit di minimum level yaitu 1000 dollar. profit yang akan didapatkan bank dari deposito berjangka diukut dari interest rate yang ditawarkan dari deposito itu berdasarkan durasi periodenya dan cost operasional dari bank. anggaplah kita menggunakan kalkulasi sederhana dengan mengasumsikan bahwa semua depositor akan melakukan deposit di jangka waktu 1 tahun. menurut sumber [disini](https://www.hsbc.co.id/1/2/!ut/p/z1/nY9NC4JAGIR_iwevvq8rSnXTgjWttg9M20tobGtgrqgl9OuLuiQERXObYZ6BAQ4J8DK9nmTanlSZFg-_485-6c8oDRiyQWSauDJxHC1YaNENgfhZGBJ_SlmIcxb4NrqTLTOXfkjQdoD_wmNPLnpr4lmIlJF_-Pelb3wAXBYqe111y8waSOC1OIpa1MalfsR521bNSEcdu64zpFKyEMZBnXX8hOSqaSHpN6E6R1Fym4nY1bQ7RRe8Og!!/dz/d5/L2dBISEvZ0FBIS9nQSEh/) rate yang digunakan untuk deposit 1 tahun adalah 3% dari nilai depositnya. jadi, dengan melakukan kalkulasi yang disederhanakan, kita bisa menghitung profit dari deposit dengan rumus :

```python
def outlierbounds(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    ub = q3 + 1.5*iqr
    lb = q1 - 1.5*iqr
    return ub, lb

a,_ = outlierbounds(df.balance)
a
```

***Insight***

- bedasarkan balance dari nasabah, kita memiliki keberagaman rentang yang sangat variatif. kita bisa lihat bahwa ketika balance sudah lebih dari kurang lebih 4019, sudah dianggap sebagai outliers. kita bisa nilai bahwa kolom balance tidak terdistribusi normal.

### Cross Validation

Cross validation yang akan dilakukan adalah dengan menggunakan stratified 5 KFold. ini berarti memfitting data ke 5 bagian berbeda dari data training dengan pembagian y yang stratified(seimbang). kita akan menggunakan 9 algoritma berbeda yang akan kita bandingkan hasilnya dari f0.5, precision, dan accuracynya.

```python
from sklearn.utils.class_weight import compute_sample_weight
sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)
```

```python
logreg = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier(class_weight = 'balanced')
rf = RandomForestClassifier(class_weight = 'balanced')
xgb = XGBClassifier()
gb = GradientBoostingClassifier()
ada = AdaBoostClassifier()
vot = VotingClassifier(estimators=[('clf1', logreg),
                             ('clf2', knn),
                             ('clf3', dt)]) 
sta = StackingClassifier(estimators=[('clf1', logreg),
                               ('clf2', knn),
                               ('clf3', dt)], 
                              final_estimator=logreg) 
```

```python
def fbeta_05_score(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    beta = 0.5  
    return (1 + beta**2) * (precision * recall) / ((beta**2 * precision) + recall)

f05_scorer = make_scorer(fbeta_05_score)
```

```python
def fbeta_2_score(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    beta = 2  
    return (1 + beta**2) * (precision * recall) / ((beta**2 * precision) + recall)

f2_scorer = make_scorer(fbeta_2_score)
```

```python
models = [logreg,knn,dt,rf,xgb,gb,ada,vot,sta]
avgf05=[]
stdf05=[]
avgpr=[]
stdfpr=[]
avgac=[]
stdfac=[]

for i in models:
    skfold=StratifiedKFold(n_splits=5)
    estimator=Pipeline([
        ('preprocess1',ct1),
        ('model',i)])
    model_cv=cross_val_score(estimator,X_train,y_train,cv=skfold,scoring=f05_scorer)
    avgf05.append(model_cv.mean())
    stdf05.append(model_cv.std())
    model_cv=cross_val_score(estimator,X_train,y_train,cv=skfold,scoring='precision')
    avgpr.append(model_cv.mean())
    stdfpr.append(model_cv.std())
    model_cv=cross_val_score(estimator,X_train,y_train,cv=skfold,scoring='accuracy')
    avgac.append(model_cv.mean())
    stdfac.append(model_cv.std())
    
crossvalres = pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost', 'GradientBoosting', 
                       'AdaBoosting', 'Voting', 'Stacking'],'mean f0.5 score':avgf05,'sdev f0.5':stdf05,
                          'mean precision':avgpr,'sdev pr':stdfpr,'mean accuracy':avgac,'sdev ac':stdfac}).set_index('model')\
                        .sort_values(by='mean f0.5 score',ascending=False)

crossvalres
```

### Best Model Explanation

***Gradient Boosting***

Gradient Boosting adalah teknik machine learning ensemble yang digunakan untuk tugas regresi dan klasifikasi. Teknik ini membangun model secara berurutan, di mana setiap model baru mencoba memperbaiki kesalahan yang dibuat oleh model sebelumnya. Tujuannya adalah untuk meningkatkan akurasi prediktif dengan menggabungkan kekuatan beberapa model, biasanya pohon keputusan.

Cara Kerja Gradient Boosting:
- Initialization: Mulai dengan model awal, sering kali model sederhana seperti memprediksi rata-rata variabel target untuk regresi atau modus untuk klasifikasi.

- Sequential Model Building: Model dibangun satu per satu. Setiap model baru dilatih untuk memprediksi kesalahan residual (perbedaan antara nilai aktual dan prediksi) dari kumpulan model sebelumnya.

- Gradient Descent: Model baru meminimalkan fungsi loss (misalnya, Mean Squared Error untuk regresi atau Log Loss untuk klasifikasi) dengan menyesuaikan terhadap gradien negatif dari fungsi loss sehubungan dengan prediksi model sebelumnya. Istilah "gradient" berasal dari sini.

- Update the Ensemble:  Prediksi model baru ditambahkan ke prediksi model sebelumnya untuk memperbarui model keseluruhan.

Gradient Boosting dipakai ketika kita menghadapi data yang non-linear dan kompleks. serta ketika kita ingin mendapati score accuracy yang tinggi. Gradient boosting juga bisa menghandle imbalance dengan menyesuaikan loss function dan class weight.

### Best Model Hyperparameter Tuning

```python
pipe_prep_model = Pipeline([
    ('preprocessing', ct1),
    ('model', gb)
])

hyperparam = {
    'model__n_estimators': range(100,1001,100),         
    'model__learning_rate': [0.001, 0.01, 0.1, 1, 10, 100],  
    'model__max_depth': range(2,10,1),               
    'model__min_samples_split': range(2,20,1),         
    'model__min_samples_leaf': range(1,20,1),         
    'model__subsample': [0.4, 0.6, 0.8, 1.0],            
    'model__max_features': ['auto', 'sqrt', 'log2']  
}

skfold=StratifiedKFold(n_splits=5)
```

***Gradient Boosting Hyperparameters:***

- n estimators: The number of boosting stages or trees to be added.
- learning rate: The step size reduction applied to each tree's contribution.
- max depth: The maximum depth of the individual trees
- min samples split:  The minimum number of samples required to split an internal node.
- min samples leaf:  The minimum number of samples required to be at a leaf node.
- subsample: The fraction of samples used for fitting individual base learners.
- max feature: The number of features to consider when looking for the best split.  

***Hyperparameter Tuning Strategy***

Disini kita akan menggunakan RandomizedSearchCV dengan 500 iteration untuk Precision dan accuracy scorer. ### Additional Model: Multilayer Perceptron

***(tidak jadi dipakai karena baseline model lebih buruk dari gradient boosting, tetapi mungkin akan bisa lebih baik lagi dengan tuning menggunakan keras tuner)***

```python
import tensorflow as tf
import keras
from keras.models import Sequential 
from keras.layers import Dense,Activation,Dropout 
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
```

```python
#reshape the y train and test to categorical encoding binary class tensor
YNN_train = to_categorical(y_train, 2)
YNN_test = to_categorical(y_test, 2)
```

```python
X_train_p.info()
```

```python
tf.keras.backend.clear_session()
model=Sequential()
model.add(Dense(256,input_dim=len(X_train_p.head(10).columns),activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(2,activation='softmax'))

#assign optimizer and compile the model
opt = tf.keras.optimizers.Adam(learning_rate=0.001) 
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=3)

#get the model architecture
model.summary()
```

```python
#train the neural network
train = model.fit(X_train_p, YNN_train, epochs=50, batch_size=64, validation_split = 0.2,callbacks=[early_stopping])
```

```python
"""this section is to plot the accuracy and loss"""
#plotting accuracy and loss
acc = train.history['accuracy']
val_acc = train.history['val_accuracy']
loss = train.history['loss']
val_loss = train.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

```python
"""this section is to evaluate the neural network model"""

#get test score and accuracy
score, acc = model.evaluate(X_test_p, YNN_test,batch_size=32)
print('Test score: %.2f'%(score*100),'%')
print('Test accuracy: %.2f'%(acc*100),'%')

#get classification report
nnpred = model.predict(X_test_p)
nnpredmax = np.argmax(nnpred,axis = 1)
print(classification_report(y_test, nnpredmax, target_names = labels))
```
